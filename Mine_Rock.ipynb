# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import logisticregression
from sklearn.metrics import accuracy_score
```

# Loading The Dataset

```python
data = pd.read_csv('DataSet.csv', header = None)  # Put the path to the dataset
```

# The Shape of Data

```python
# Rows and Columns
data.shape
```

# Data Description

```python
data.describe()
```

```python
# Rocks and Mines Quantity
data[60].value_counts()
```

# Mean Value for Rocks and Mines

```python
data.groupby(60).mean()
```

# Separating Data and Label

```python
X = data.drop(columns = 60, axis = 1)  # All the data is being stored in X except for the last column
Y = data[60]                           # The last column is stored in Y
```

```python
print(X)
print(Y)
```

# Train and Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
```

# Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

# Evaluation

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the test data : ', test_data_accuracy)
```
